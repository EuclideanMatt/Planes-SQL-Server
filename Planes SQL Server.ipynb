```python
"""

Here I will simply start this planes project by creating a SQL Server database and import three files that will be used to complete our task:
1.) planes.xlx
2.) latitudes and longitudes 
3.) City names

"""
```

```python
# packages 

import sqlite3
import pandas as pd 
```

```python
# creating a new database to work in

conn = sqlite3.connect(r"C:\Users\Epicm\Desktop\data_bases\dbs\planes\plane1.db")
cursor = conn.cursor()
```

```python
## creating dataframes of all of our desired data

# dataset of flight
planes_df = pd.read_excel(r"C:\Users\Epicm\Desktop\data_bases\descriptive_stats_exploratory_anlysis\data\flight_train.xlsx")

# dataset of latitude and longitudes
locations_df = pd.read_csv(r"C:\Users\Epicm\Desktop\data_bases\descriptive_stats_exploratory_anlysis\data\ind_loc.csv")

# dataset of city names

cities_df = pd.read_csv(r"C:\Users\Epicm\Desktop\data_bases\descriptive_stats_exploratory_anlysis\data\indian_city_names.csv")
```

```python
# now lets build out all our desired datasets to import as blank tables

# main planes data 

cursor.execute('''
CREATE TABLE IF NOT EXISTS flights (
    airline TEXT,
    date_of_journey TEXT,  -- Store as TEXT in 'YYYY-MM-DD' format
    source TEXT,
    destination TEXT,
    route TEXT,
    dep_time TEXT,         -- Store time as TEXT in 'HH:MM' format
    arrival_time TEXT,     -- Store time as TEXT in 'HH:MM' format
    duration TEXT,
    total_stops TEXT,
    additional_info TEXT,
    price INTEGER
)
''')


for _, row in planes_df.iterrows():
    cursor.execute('''
    INSERT INTO flights (airline, date_of_journey, source, destination, route, dep_time, arrival_time, duration, total_stops, additional_info, price)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', row)
    
conn.commit()

# latitudes and longitude data
cursor.execute('''
CREATE TABLE IF NOT EXISTS locations (
   city TEXT PRIMARY KEY,
   latitude FLOAT,
   longitude FLOAT
)
''')

for _, row in locations_df.iterrows():
    cursor.execute('''
    INSERT INTO locations (city, latitude, longitude )
    VALUES (?, ?, ?)
    ''', row)
    
conn.commit()



# city names data 
cursor.execute('''
CREATE TABLE IF NOT EXISTS cities (
   abbreviation TEXT PRIMARY KEY,
   full_name TEXT
   
)
''')

for _, row in cities_df.iterrows():
    cursor.execute('''
    INSERT INTO cities (abbreviation, full_name )
    VALUES (?, ?)
    ''', row)
    
conn.commit()



```

```python
# let's check to make sure our data imported correctly 
```

```python
# flights

cursor.execute('SELECT * FROM flights')

# Fetch and print the results
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# locations

cursor.execute('SELECT * FROM locations')

# Fetch and print the results
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# cities 

cursor.execute('SELECT * FROM cities')

# Fetch and print the results
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
# all the tables were created successfully
```

```python
# closing connection to database and cursor
cursor.close()
conn.close()
```
